```python
import pandas as pd
import pickle
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def open_contact_and_mse_file(contact_path, mse_path):
    with open(contact_path, "rb") as f:
        contact = pickle.load(f)
    mse = pd.read_csv(mse_path)
    return contact, mse["MSE(mean)"]
```

```python
def get_density_from_contact(contact):
    density = {i:None for i in contact.keys()}
    for i in contact.keys():
        num = 0
        for j in contact[i].keys():
            num += contact[i][j].__len__()
        num = num / 2
        keys = len(list(contact[i].keys()))
        keys = keys * (keys - 1) / 2
        if keys == 0:
            density[i] = 0
        else:
            density[i] = num / keys
    return density
```

```python
def get_density_and_mse(density, mse):
    xpoints = [density[i] for i in range(1, density.__len__()+1)]
    xy = dict()
    idx = 0
    for x in xpoints:
        if x == 0:
            idx += 1
            continue
        xy[x] = mse[idx] 
        idx += 1
    xy_x = sorted(list(xy.keys()))
    xy_y = [xy[i] for i in xy.keys()]
    return xy_x, xy_y
    
    
```

```python
def get_x_and_y(contact_path, mse_path):
    contact, mse = open_contact_and_mse_file(contact_path, mse_path)
    density = get_density_from_contact(contact)
    x, y = get_density_and_mse(density, mse)
    return x, y, density, mse
    
```

```python
contact_50 = "contact_7474.pkl"
contact_100 = "contact_8628.pkl"
contact_150 = "contact_9440.pkl"
contact_200 = "contact_10000.pkl"
contact_250 = "contact_10369.pkl"
contact_350 = "contact_10873.pkl"
mse_50 = "2022_05_03_17_59_22_NM50.csv"
mse_100 = "2022_05_03_18_35_00_NM100.csv"
mse_150 = "2022_05_03_19_15_37_NM150.csv"
mse_200 = "2022_05_03_20_00_24_NM200.csv"
mse_250 = "2022_05_03_20_47_04_NM250.csv"
mse_350 = "2022_05_03_21_36_27_NM350.csv"
```

```python
x50, y50, density50, mse50= get_x_and_y(contact_50, mse_50)
x100, y100, density100, mse100 = get_x_and_y(contact_100, mse_100)
x150, y150, density150, mse150 = get_x_and_y(contact_150, mse_150)
x200, y200, density200, mse200 = get_x_and_y(contact_200, mse_200)
x250, y250, density250, mse250 = get_x_and_y(contact_250, mse_250)
x350, y350, density350, mse350 = get_x_and_y(contact_350, mse_350)
```

```python
plt.xlabel("density")
plt.ylabel("mse(mean)")
plt.title("relation between density and mse when the stride is 50m")
plt.plot(x50, y50, marker='.')
plt.show()
plt.xlabel("density")
plt.ylabel("mse(mean)")
plt.title("relation between density and mse when the stride is 50m removing outliers")
plt.plot(x50[4:-3], y50[4:-3], marker='.')
plt.show()
```
